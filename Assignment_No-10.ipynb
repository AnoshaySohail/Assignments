# 1.	List all cities where the area is greater than 5000, along with their population.

```python
sql_query("""SELECT 
    cities.name AS city_name,
    cities.population
FROM 
    cities
JOIN 
    facts
ON 
    cities.facts_id = facts.id
WHERE 
    facts.area > 5000;""")
```

# 2.	List the names of capital cities where the death rate is greater than 2% and the birth rate is less than 20%

```python
sql_query("""SELECT 
    cities.name AS capital_city
FROM 
    cities
JOIN 
    facts
ON 
    cities.facts_id = facts.id
WHERE 
    cities.capital = 1
    AND facts.death_rate > 2
    AND facts.birth_rate < 20;""")
```

# 3.	List the cities with a population growth rate between 1% and 5%, sorted in desc by population growth.

```python
sql_query("""SELECT 
    cities.name AS city_name,
    facts.population_growth
FROM 
    cities
JOIN 
    facts
ON 
    cities.facts_id = facts.id
WHERE 
    facts.population_growth BETWEEN 1 AND 5
ORDER BY 
    facts.population_growth DESC;""")
```

# 4.	Find countries that start with "A" and do not have any cities listed in the cities table. The query should return the country name and city name. Using left join

```python
sql_query("""SELECT 
    facts.name AS country_name,
    cities.name AS city_name
FROM 
    facts
LEFT JOIN 
    cities
ON 
    facts.id = cities.facts_id
WHERE 
    facts.name LIKE 'A%'
    AND cities.id IS NULL;""")
```

# 5.	List all capital cities and their corresponding death rate using a LEFT JOIN.

```python
sql_query("""SELECT 
    cities.name AS capital_city,
    facts.death_rate
FROM 
    cities
LEFT JOIN 
    facts
ON 
    cities.facts_id = facts.id
WHERE 
    cities.capital = 1;""")
```

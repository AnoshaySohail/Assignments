```python
import sqlite3 as sql
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#connection
conn = sql.connect('database.sqlite')
conn
```

```python
q = "SELECT * FROM Salaries"
pd.read_sql_query(q, conn)
```

```python
df = pd.read_sql_query(q, conn)
df = df[(df['JobTitle'].str.lower() != 'not provided') & (df['EmployeeName'].str.lower() != 'not provided')]
```

```python
df.head()
```

# 1. What is the average TotalPayBenefits for each JobTitle? exclude ('Not Provided', 'Not provided')

```python
avg_totalpay_by_job = df.groupby('JobTitle')['TotalPayBenefits'].mean().reset_index()
avg_totalpay_by_job
```

# 2. Which JobTitle has the most employees in each year? exclude ('Not Provided', 'Not provided')

```python
most_common_job_each_year = df.groupby(['Year', 'JobTitle']).size().reset_index(name='Count')
most_common_job_each_year = most_common_job_each_year.loc[most_common_job_each_year.groupby('Year')['Count'].idxmax()]
most_common_job_each_year
```

# 3. What is the total number of unique employees? exclude ('Not Provided', 'Not provided')

```python
unique_employees_count = df['EmployeeName'].nunique()
unique_employees_count
```

# 4. Show All Employees Ordered By Their TotalPayBenefits In Descending Order? exclude ('Not Provided', 'Not provided')

```python
employees_sorted_by_pay = df[['EmployeeName', 'JobTitle', 'TotalPayBenefits']].sort_values(by='TotalPayBenefits', ascending=False)
employees_sorted_by_pay.head()
```

# 5. Show All Employees With A TotalPaybenefits Value Between 125,000 and 150,000 And A Job Title 'Firefighter'? exclude ('Not Provided', 'Not provided')

```python
firefighters_filtered = df[(df['TotalPayBenefits'].between(125000, 150000)) & (df['JobTitle'].str.lower() == 'firefighter')]
firefighters_filtered
```

# 6. How many employees have no overtime pay or other pay? exclude ('Not Provided', 'Not provided')

```python
no_overtime_count = df[(df['OvertimePay'] == 0) & (df['OtherPay'] == 0)].shape[0]
no_overtime_count
```

```python
num_cols = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay', 'TotalPayBenefits']
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors='coerce')
```

# 7. Show top 10 employees having (overtimepay+OtherPay) in bar plot?  exclude ('Not Provided', 'Not provided')

```python
df['ExtraPay'] = df['OvertimePay'] + df['OtherPay']
top_10_extra_pay = df[['EmployeeName', 'ExtraPay']].nlargest(10, 'ExtraPay')

plt.figure(figsize=(12, 6))
plt.barh(top_10_extra_pay['EmployeeName'], top_10_extra_pay['ExtraPay'], color='skyblue')
plt.xlabel('OvertimePay + OtherPay')
plt.ylabel('Employee Name')
plt.title('Top 10 Employees with Highest Extra Pay (OvertimePay + OtherPay)')
plt.gca().invert_yaxis()  # Invert y-axis to show highest at the top
plt.show()
```

# 8. Show The Average of BasePay, OverTime, OtherPay for all employees in pie chart? exclude ('Not Provided', 'Not provided')

```python
avg_values = df[['BasePay', 'OvertimePay', 'OtherPay']].mean()

plt.figure(figsize=(8, 8))
plt.pie(avg_values, labels=avg_values.index, autopct='%1.1f%%', colors=['lightblue', 'orange', 'lightgreen'])
plt.title('Average Pay Distribution (BasePay, OvertimePay, OtherPay)')
plt.show()
```

# 9. Get the average TotalPayBenefits for each JobTitle, but only for those with an average pay greater than 75000? exclude ('Not Provided', 'Not provided')

```python
high_avg_totalpay = df.groupby('JobTitle')['TotalPayBenefits'].mean().reset_index()
high_avg_totalpay = high_avg_totalpay[high_avg_totalpay['TotalPayBenefits'] > 75000]

high_avg_totalpay
```

# 10. Get the number of employees (EmployeeName) for each JobTitle, where the count is greater than 10?  exclude ('Not Provided', 'Not provided')

```python
job_title_counts = df.groupby('JobTitle')['EmployeeName'].count().reset_index()

job_title_counts = job_title_counts[job_title_counts['EmployeeName'] > 10]

job_title_counts
```
